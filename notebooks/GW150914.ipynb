```python
from gwpy.timeseries import TimeSeries

# GW150914 event time
start = 1126259446
end = 1126259478

# Get data from LIGO Hanford detector
data = TimeSeries.fetch_open_data("H1", start, end)
```

```python
data.plot()
```

```python
filtered = data.bandpass(20, 300)
filtered.plot()
```

```python
white = filtered.whiten()
white.plot()
```

```python
white.crop(1126259460, 1126259470).plot()
```

```python
white.q_transform().plot()
```

```python
import numpy as np

peak_time = white.time[np.argmax(abs(white.value))]
print(peak_time)
```

```python
snr = abs(white).max() / np.std(white.value) #signal to noise ratio
print(snr)
```
