```python
from gwpy.timeseries import TimeSeries

# GW150914 = first detected black hole merger
data = TimeSeries.fetch_open_data("H1", 1126259446, 1126259478)

data.plot()
```

```python
clean = data.whiten().bandpass(50, 250)

clean.plot()
```

```python
q = clean.q_transform(frange=(20, 500))
q.plot()
```

```python
clean.crop(1126259462, 1126259466).plot()
```

```python
event = clean.crop(1126259460, 1126259466)

event.plot()
```
